All in One

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2, 1000, 100000)

# Function 1 - Linear time complexity
y1 = x

# Function 2 - Logarithmic time complexity
y2 = np.log(x)

# Function 3 - Superlinear time complexity
y3 = y= x * np.log(np.log(x))

# Function 4 - Superlinear time complexity
y4 = x * np.log(x)

# Function 5 - Quadratic time complexity
y5 = x ** 2

# Function 6 - Quadratic time complexity
y6 = x ** 2 + x

# Plot all functions
plt.plot(x, y1, label='x')
plt.plot(x, y2, label='log(x)')
plt.plot(x, y3, label='x * log(log(x))')
plt.plot(x, y4, label='x * log(x)')
plt.plot(x, y5, label='x^2')
plt.plot(x, y6, label='x^2 + x')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Various Time Complexities')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2, 10, 10000)   #(2, 10, 10000)

# Function 1 - Linear time complexity
y1 = x

# Function 2 - Logarithmic time complexity
y2 = np.log(x)

# Function 3 - Superlinear time complexity
y3 = y= x * np.log(np.log(x))

# Function 4 - Superlinear time complexity
y4 = x * np.log(x)

# Function 5 - Quadratic time complexity
y5 = x ** 2

# Function 6 - Quadratic time complexity
y6 = x ** 2 + x

# Plot all functions
plt.plot(x, y1, label='x')
plt.plot(x, y2, label='log(x)')
plt.plot(x, y3, label='x * log(log(x))')
plt.plot(x, y4, label='x * log(x)')
plt.plot(x, y5, label='x^2')
plt.plot(x, y6, label='x^2 + x')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Various Time Complexities')
plt.legend()
plt.grid(True)
plt.show()
```
